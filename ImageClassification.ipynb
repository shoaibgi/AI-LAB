```python
import cv2
import os

def video_to_images(video_path, output_folder, prefix="frame", image_format="jpg"):
    # Create output directory if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Open the video file
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break  # No more frames to read

        # Construct image file name
        filename = f"{prefix}_{frame_count:05d}.{image_format}"
        filepath = os.path.join(output_folder, filename)

        # Save the frame as an image
        cv2.imwrite(filepath, frame)

        frame_count += 1

    cap.release()
    print(f"Saved {frame_count} frames to {output_folder}")

# Example usage
video_path = "robot.mp4"
output_folder = "D:\Training data\Robot"
video_to_images(video_path, output_folder)
```
